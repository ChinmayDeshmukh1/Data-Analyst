```python
1. Entering for loop for each item in sequence > Last item reached > Yes > Exiting Loop > Statement after the for loop
2. Entering for loop for each item in sequence > Last item reached > No > body of the loop > Last element reached?
```

```python
integers = [1,2,3,4,5]
```

```python
for number in integers:
    print(number)
```

```python
for number in integers:
    print('yep!')
```

```python
for Jelly in integers:
    print(Jelly + Jelly)
```

```python
ice_cream_dict = {'name': 'Alex Freberg', 'weekly intake': 5, 'favorite ice cream': ['MCC', 'Chocolate']}
```

```python
for cream in ice_cream_dict.values():
    print(cream)
```

```python
for key, value in ice_cream_dict.items():
    print(key, '>',value)
```
