```python
While Loops
```

```python
1. Enter the while loop > Test Condition > False > Exit the while loop
2. Enter the while loop > Test Condition > True > while loop body > Test Condition
```

```python
number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No longer < 5')
```

```python
number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)
else:
    print('No longer < 5')
```
