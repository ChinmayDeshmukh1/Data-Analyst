```python
file_name = os.listdir(path)
```

```python
for file in file_name:
     if ".csv" in file and not os.path.exists(path + "csv files/" + file):
         shutil.move(path + file, path + "csv files/" + file)
     elif ".png" in file and not os.path.exists(path + "image files/" + file):
         shutil.move(path + file, path + "image files/" + file)
     elif ".txt" in file and not os.path.exists(path + "text files/" + file):
         shutil.move(path + file, path + "text files/" + file)
     else:
         print("There are files in this path that were not moved!")
```

```python
from pathlib import Path
import shutil

# 1. Define your base directory as a Path object
base_path = Path(r"D:\SQL\21. Pivot Tables in Excel")

# 2. Iterate through files in the directory
for file_path in base_path.iterdir():
    if file_path.is_file():
        file_name = file_path.name
        
        # Check for .csv files
        if ".csv" in file_name:
            dest_dir = base_path / "csv files"
        # Check for .png files
        elif ".png" in file_name:
            dest_dir = base_path / "image files"
        # Check for .txt files
        elif ".txt" in file_name:
            dest_dir = base_path / "text files"
        else:
            print(f"There are files in this path that were not moved: {file_name}")
            continue

        # 3. Create destination folder if it doesn't exist
        dest_dir.mkdir(exist_ok=True)
        
        # 4. Define full target file path
        target_file = dest_dir / file_name
        
        # 5. Move the file if it doesn't already exist at the destination
        if not target_file.exists():
            shutil.move(str(file_path), str(target_file))
```
